```python
import pandas as pd #Helps with data analysis

import numpy as np #Deals with multidimentional arrays

data = pd.read_csv("dataset.csv") #Gets data from the csv file

data.head()
```

```python
data.isnull().sum() #Tells us which values are 0
```

```python
data.type.value_counts() #Tells us number of transaction types (5 different ones)
```

```python
type = data["type"].value_counts()

transactions = type.index #takes the name, x axis

quantity = type.values #takes the count (how many of each type), y axis

import plotly.express as px

figure = px.pie(data, values=quantity, names=transactions, hole=0.3, title="Distribution of Data Types")

figure.show()
```

```python
numeric_data = data.select_dtypes(include=[np.number]) #convert non numeric data into numeric data

correlation = numeric_data.corr()
print(correlation["isFraud"].sort_values(ascending=False))
```

```python
data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2, "CASH_IN": 3, "TRANSFER": 4, "DEBT": 5})

data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data.head()
```

```python
#Use significant values as x values (for example, don't use nameOrig), and Fraud/NotFraud is y output

from sklearn.model_selection import train_test_split

x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]]) #splitting the data (x)

y = np.array(data[["isFraud"]]) #splitting the data (y)

from sklearn.tree import DecisionTreeClassifier

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=42) #splits 80 to 20 %
model = DecisionTreeClassifier()

model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

```python
features= np.array([[4, 9000.60, 9000.60, 0.0]])
print(model.predict(features))
```
